```python
import numpy as np

# Generate a dataset of numbers from 1 to 100
data = np.arange(1, 101, 1)

# Compute statistics
mean_val = np.mean(data)
median_val = np.median(data)
variance_val = np.var(data)
std_dev_val = np.std(data);

# Display the results
print(f"Generated dataset: {data}")
print(f"Mean: {mean_val:.2f}")
print(f"Median: {median_val:.2f}")
print(f"Variance: {variance_val:.2f}")
print(f"Standard Deviation: {std_dev_val:.2f}")
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

# The 'data' variable is assumed to be defined from the previous cell (numbers from 1 to 100).
# If this cell is run independently, uncomment the line below:
# data = np.arange(1, 101, 1)

plt.figure(figsize=(10, 6))
sns.kdeplot(data, fill=True, color='purple', linewidth=2)
plt.title('Probability Density Estimate of Numbers from 1 to 100')
plt.xlabel('Number Value')
plt.ylabel('Density')
plt.xticks(np.arange(0, 101, 10))
plt.grid(True, linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Define parameters for the normal distribution
mu = 50       # Mean as requested by the user
sigma = 15    # Standard deviation chosen so most data falls roughly between 1 and 100
sample_size = 10000 # Number of data points

# Generate data for a normal distribution
normal_data = np.random.normal(mu, sigma, sample_size)

# Plotting the histogram
plt.figure(figsize=(10, 6))
plt.hist(normal_data, bins=50, density=True, alpha=0.7, color='darkgreen', edgecolor='black')
plt.title('Histogram of a Normal Distribution (Mean=50, Std Dev=15)')
plt.xlabel('Value')
plt.ylabel('Probability Density')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

```python
import numpy as np

# --- Coin Toss Simulation ---

# Number of coin tosses to simulate
num_tosses = 10000

# Simulate coin tosses (0 for Tails, 1 for Heads)
tosses = np.random.randint(0, 2, size=num_tosses)

# Count occurrences of Heads (1) and Tails (0)
heads_count = np.sum(tosses == 1)
tails_count = np.sum(tosses == 0)

# Compute probabilities
prob_heads = heads_count / num_tosses
prob_tails = tails_count / num_tosses

# Display results
print(f"--- Coin Toss Experiment Results ({num_tosses} tosses) ---")
print(f"Number of Heads: {heads_count}")
print(f"Number of Tails: {tails_count}")
print(f"Probability of Heads: {prob_heads:.4f}")
print(f"Probability of Tails: {prob_tails:.4f}")
```
