```python
matrix = []

print("Enter the elements of the 3x3 matrix row-wise:")
for i in range(3):
    row = list(map(int, input(f"Enter row {i+1}: ").split()))
    matrix.append(row)

print("\nMatrix:")
for row in matrix:
    print(row)

# Assign elements
a, b, c = matrix[0]
d, e, f = matrix[1]
g, h, i = matrix[2]

# Determinant
det = a*(e*i - f*h) - b*(d*i - f*g) + c*(d*h - e*g)

print("\nDeterminant =", det)

if det == 0:
    print("Inverse does not exist (determinant is zero).")
else:
    # Cofactor matrix
    cofactor = [
        [(e*i - f*h), -(d*i - f*g), (d*h - e*g)],
        [-(b*i - c*h), (a*i - c*g), -(a*h - b*g)],
        [(b*f - c*e), -(a*f - c*d), (a*e - b*d)]
    ]

    # Adjoint (transpose of cofactor)
    adjoint = [
        [cofactor[0][0], cofactor[1][0], cofactor[2][0]],
        [cofactor[0][1], cofactor[1][1], cofactor[2][1]],
        [cofactor[0][2], cofactor[1][2], cofactor[2][2]]
    ]

    # Inverse matrix
    inverse = [[adjoint[i][j] / det for j in range(3)] for i in range(3)]

    print("\nInverse of the matrix:")
    for row in inverse:
        print(row)
```

```python
import numpy as np

print("Enter coefficients for equation 1 (a b):")
a1, b1 = map(float, input().split())

print("Enter constant term for equation 1:")
c1 = float(input())

print("\nEnter coefficients for equation 2 (a b):")
a2, b2 = map(float, input().split())

print("Enter constant term for equation 2:")
c2 = float(input())

# Coefficient matrix
A = np.array([[a1, b1],
              [a2, b2]])

# Constant matrix
B = np.array([c1, c2])

# Solving the equations
solution = np.linalg.solve(A, B)

print("\nSolution:")
print("x =", solution[0])
print("y =", solution[1])
```
